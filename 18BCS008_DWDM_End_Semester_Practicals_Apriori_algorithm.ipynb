# Apriori algorithm

### Dataset: Groceries dataset for Market Basket Analysis(MBA) (basket.csv)

```python
#Importing the packages

import pandas as pd
from apyori import apriori
```

```python
#Reading the dataset

data=pd.read_csv("basket.csv")
data
```

```python
#Dataset description
#Number of rows and columns

data.shape
```

```python
#Dataset description

data.describe()
```

```python
#Attributes and data types

data.info()
```

```python
#Converting dataset into list

records_list=[]
for i in range(0,14963):
    records_list.append([str(data.values[i,j]) for j in range(0,11)])
```

```python
#Transaction sets

for i in records_list:
    print(i)
```

#### min support=0.5

```python
#Apriori algorithm

association_rules1=apriori(records_list,min_support=0.5,min_confidence=0.7,min_length=2)
```

```python
#Converting into list of association rules

association_rules_results1=list(association_rules1)
```

```python
#Number of association rules

print(len(association_rules_results1))
```

```python
association_rules_results1[0]
```

```python
#List of rules

for i in association_rules_results1:
    j=i[0]
    n=[m for m in j]
    print("Rule :"+str(i[0])+"->"+str(i[1]))
    print("Support : {}".format(i[1]))
    print("Confidence : {}".format(i[2][0][2]))
    print("Lift : {}".format(i[2][0][3]))
    print("\n------------------------------------------------\n")
```
